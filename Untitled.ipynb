## dataframe_dict：key为日期，value为当天所有成交记录的dataframe组成的列表

```python
dataframe_dict={}

def finding_nemo_1(String,Substr,times): 
    '''
    在String中查找Substr出现第times次的索引。
    '''
    #s = time.time()
    nemo = 0
    for i in range(1,times+1):
        nemo = String.find(Substr,nemo) + 1
        if nemo == 0:
            return -1
    return nemo-1

for date in path_dict.keys():
    dataframe_dict[date]=[]
    for path in path_dict[date]:
        df=0
        try:
            df=pd.read_csv(path,index_col=False)
        except:
            continue
        finally:
            if isinstance(df,int):
                continue
            #print(df)
            start=finding_nemo_1(path,"/",3)
            end=finding_nemo_1(path,"/",4)
            column_name=list(df.columns)
            for i in range(len(column_name)):
                column_name[i]=column_name[i].replace("\r", " ")
                column_name[i]=column_name[i].replace("\n", "")
            df.columns = column_name
            df.name=path[start+1:end]
            dataframe_dict[date].append(df)

for date in dataframe_dict.keys():
    tmp1=dataframe_dict[date]
    for i in range(len(tmp1)):
        m,n=tmp1[i].shape
        columns_check=["成交金额 Turnover","手续费 Fee","成交量 Lots"]
        for column in columns_check:
            for j in range(m):
                if isinstance(tmp1[i][column][j],str):
                    tmp1[i].loc[j,column]=float(list(tmp1[i][column])[j].replace(",",""))
```
